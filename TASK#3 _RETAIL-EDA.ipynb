### Checking correlation between columns visually

```python
f,ax = plt.subplots(figsize=(12,8))
corr_mat = df.corr()
sns.heatmap(corr_mat,xticklabels=corr_mat.columns,yticklabels=corr_mat.columns,annot=True)
print(corr_mat)
```

No such strong correlation is seen from above heatmap.

Hence, we can derive that the correlation between "Profit" and "Sales" is "0.4791". This shows that there is a positive realtion between them.

Correlation between "Profit" and "Discount" is "-0.2197". This means that there is a weak negative relation between both of these numeric features.

### Checking for outliers   

```python
plt.figure(figsize=(12,8))
df_val = df[['Sales','Quantity','Discount','Profit']].copy()
sns.set_theme(style="whitegrid")
sns.boxplot(data=df_val,width=1)
```

From the above visualization,the presence of outliers are clearly seen in Sales and Profit columns.

```python
#Removal of Outliers
df_val =df_val[(np.abs(stats.zscore(df_val)) < 3).all(axis=1)]
```

```python
plt.figure(figsize=(12,8))
sns.set_theme(style="whitegrid")
sns.boxplot(data=df_val,width=1)
```

Hence most of the outliers are removed from the dataset.

### Exploratory Data Analysis

```python
df.iplot(x='Region',y='Sales',kind='bar',title='Region VS Sales',xTitle='Region',yTitle='Sales')
```

From the above visualization,it's oberserved that the West leads in Sales followed by East,Central,South.

```python
plt.figure(figsize=(12,8))
sns.countplot(x='Region',data=df)
```

This shows that most of the entries are form the "West" Region of United States.

```python
plt.figure(figsize=(12,8))
sns.countplot(x='Ship Mode',data=df,palette='Reds')
```

Most of the transactions were shipped under Standard Class.With least count of Same Day delivery.

```python
plt.figure(figsize=(12,8))
sns.countplot(x='Category',data=df)
```

Most of the categories of data in the dataset are from "Office Supplies".

```python
df.groupby('Category')['Sales', 'Profit'].sum().plot.bar()
```

Above graph shows that the Technological Product Cateogory has highest sales and profit but Furniture has incurred loss despite of having greater sales than Office Supplies.

```python
df.groupby('Sub-Category')['Sales'].sum().sort_values(ascending=False).plot.bar()
```

We can see here that the Phones and chairs are the highest sold items

```python
plt.figure(figsize=(12,8))
sns.countplot(x=df['State'],order=(df['State'].value_counts().head(20)).index)
plt.xticks(rotation=90)
```

Maximum number of transaction are from people from "California" followed by "New York".
